# 一、基础部分

```python
#若爬取百度新闻的热搜新闻词 包含属性为“href”和“title”的数据

from bs4 import BeautifulSoup
from urllib.request import urlopen

#对于中文网站，需要使用decode()
html = urlopen("https://news.baidu.com").read().decode('utf-8')

soup = BeautifulSoup(html, features='lxml')
#ul的class选择
# 热搜新闻词 hotwords clearfix
# 热点要闻 ulist focuslistnews

#用于记录爬虫获取数据的条数
i = 1
f = open('baidu_news.txt','w')
for sp in soup.find_all('ul'):
    for tag in sp.find_all("a"):  
            url = tag.get("href")
            title = tag.string
            if tag.get("title") != None:
                title = tag.get("title")
            target = tag.get("target")
            if target is None:
                continue
            if title is None:
                continue
            
            #if target == "_blank": # and X in tile
            print(str(i)+":")
            print("词条："+title)
            print("链接："+url)
            f.write(str(i) + ":")
            f.write('\n')
            f.write("词条：" + title)
            f.write('\n')
            f.write("链接：" + url)
            f.write('\n')
            i = i +1
f.close()
                
```

```python
# 若爬取百度新闻的热搜新闻词 包含属性为“href”和“title”的数据

from bs4 import BeautifulSoup
from urllib.request import urlopen

# 对于中文网站，需要使用decode()
html = urlopen("https://news.baidu.com").read().decode('utf-8')

soup = BeautifulSoup(html, features='lxml')
# ul的class选择
# 热搜新闻词 hotwords clearfix
# 热点要闻 ulist focuslistnews

# 用于记录爬虫获取数据的条数
i = 1

f = open('baidu_news_keyword.txt','w')
temp = input("请输入关键词：")
for sp in soup.find_all('ul'):
    for tag in sp.find_all("a"):
        url = tag.get("href")
        title = tag.string
        target = tag.get("target")
        # if title is None:
        #    continue
        if target is None:
            continue
        if title is None:
            continue

        if  temp in title:  # and X in tile
            print(str(i) + ":")
            print("词条：" + title)
            print("链接：" + url)
            f.write(str(i) + ":")
            f.write('\n')
            f.write("词条：" + title)
            f.write('\n')
            f.write("链接：" + url)
            f.write('\n')
            i = i + 1

f.close()
```
